```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_countries = pd.read_csv('dates_countries_001.txt', sep=",")
all_countries.head()
all_countries.drop(['Lat', 'Long', 'Province/State'], axis=1,inplace=True) #country ne jer moram znati koja je koja
all_countries = all_countries.transpose()
all_countries.columns = all_countries.iloc[0]
all_countries = all_countries[1:]
all_countries.columns = ['Austria_1', 'Bosnia and Herzegovina', 'Croatia', 'France', 'Germany','Italy', 'Slovenia', 'Spain', 'United Kingdom', 'Hungary','Austria_2']
#austria mi se pojavljuje dva puta???
all_countries.drop(['Austria_2'],axis=1,inplace=True)
all_countries
all_countries.columns = ['Austria', 'Bosnia and Herzegovina', 'Croatia', 'France', 'Germany','Italy', 'Slovenia', 'Spain', 'United Kingdom', 'Hungary',]
all_countries
```

```python
all_countries = all_countries.loc[:, all_countries.columns != 'Country/Region'].diff(axis=0) #bas dnevni podaci
all_countries.drop(['1/22/20'], axis=0,inplace=True)
```

```python
all_countries #ovo su bas dnevni podaci sada
```

```python
all_countries = all_countries.set_index(pd.to_datetime(all_countries.index)) #ovaj dio je jako bitan i puno sam se mucila  s tim
#ugl fora je da sam nasla taj resample dolje koji sumira dane u tjednu, al je bio problem da radi s indeksom tim nekim posebnom
#index nije ko gora neg bas mora biti datetime tip, inace ne radi
#uoci kak je ovo kul jer 655/7 je priblizno 93, on zaokruzi na vise jer valjda gleda zadnji tjedan, nema vise onih nanova
all_countries.index
all_countries = all_countries.resample('W').sum()
```

```python
all_countries = all_countries/7 # i sad da dobim prosjeke samo podjelim sa 7
all_countries
```

```python
df = all_countries
binInterval = [-1, 100, 300,700, 2000, 1000000]
binLabels   = ['A', 'B', 'C','D','E']
df['Austria_'] = pd.cut(all_countries['Austria'], bins = binInterval, labels=binLabels)
df['Bosnia and Herzegovina_'] = pd.cut(all_countries['Bosnia and Herzegovina'], bins = binInterval, labels=binLabels)
df['Croatia_'] = pd.cut(all_countries['Croatia'], bins = binInterval, labels=binLabels)
df['France_'] = pd.cut(all_countries['France'], bins = binInterval, labels=binLabels)
df['Germany_'] = pd.cut(all_countries['Germany'], bins = binInterval, labels=binLabels)
df['Italy_'] = pd.cut(all_countries['Italy'], bins = binInterval, labels=binLabels)
df['Slovenia_'] = pd.cut(all_countries['Slovenia'], bins = binInterval, labels=binLabels)
df['Spain_'] = pd.cut(all_countries['Spain'], bins = binInterval, labels=binLabels)
df['United Kingdom_'] = pd.cut(all_countries['United Kingdom'], bins = binInterval, labels=binLabels)
df['Hungary_'] = pd.cut(all_countries['Hungary'], bins = binInterval, labels=binLabels)
df.drop(['Austria','Bosnia and Herzegovina','Croatia','France','Germany','Italy','Slovenia','Spain','United Kingdom','Hungary'],axis=1,inplace=True)
```
